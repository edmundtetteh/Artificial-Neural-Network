# unseen data generalization with the network (ResNet)
1. Transformation pipeline use to set values of image to sit well with output of 1000 ImageNet classes
2. involves scale input image (resize) to 256 x 256 px
3. center input image by cropping to 224 x 224 px
4. put it into tensor
5. Normalize the RGB to have mean and sd

```python
from torchvision import transforms
preprocess = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )])
```

```python
from PIL import Image
img = Image.open("data/p1ch2/bobby.jpg")
```

```python
img.show()
```

```python
# apply the preprocess pipeline function
img_t = preprocess(img)

```

```python
batch_t = torch.unsqueeze(img_t, 0)
```

```python
# inference
# before inference we put the network to eval mode (this enable batch normalization and dropout to work well)
resnet.eval()
```

```python
# score of the parameters
out = resnet(batch_t)
out
```

```python
# number datapoints
out.numel()
```

```python
# list of labels in same order presented in the neural network
with open('data/p1ch2/imagenet_classes.txt') as f:
    labels = [line.strip() for line in f.readlines()]
    print(labels[207])
```

```python
# index of max score
_, index = torch.max(out, 1)
index
```

```python
max_index = torch.argmax(out, dim=None)
max_index
```

```python
# use softmax to predict in the range of 0,1
percentage = torch.nn.functional.softmax(out, dim=1)[0] * 100
labels[index.item()], percentage[index.item()].item()
```

```python
# sort the result for best 5
_, indices = torch.sort(out, descending=True)
[(labels[idx], percentage[idx].item()) for idx in indices[0][:5]]
```
